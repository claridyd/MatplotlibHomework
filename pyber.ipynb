```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Read the city and ride data
city_df = pd.read_csv("city_data.csv")
ride_df = pd.read_csv("ride_data.csv")

# City dataframe
#city_df

# Ride dataframe
#ride_df

#combine the data into a single dataset
combo_data = pd.merge(ride_df, city_df, on="city")

#Display the combo data table for preview
combo_data
```

```python
# what are the city types
city_type = combo_data["type"].unique()
#print(city_type)

#Group by City
city_grp = combo_data.groupby("city")


#Total number of rides per city
rides_per_city = city_grp["ride_id"].count()
#print(rides_per_city)

#Total number of drivers per city
drivers_city = city_grp["driver_count"].max()
#print(drivers_city)

#list city type
city_type = city_grp["type"].max()
#print(city_type)

#Average Fare per city
avg_fare = city_grp["fare"].mean()
#print(avg_fare)
#######################################################
# Create dataframe to hold data
#zipped_df = list(zip(rides_per_city, drivers_city, avg_fare))
#test1 = pd.DataFrame(zipped_df, columns = ["Rides per City", "Drivers per City", "Average Fare"])
#print(test1)

city_summary = pd.DataFrame({"Rides per City": rides_per_city,
                            "Drivers per City": drivers_city,
                            "Average Fare": avg_fare,
                            "City Type": city_type})
city_summary.head()

```

```python
#Get x, y for the city types
#Urban 
urban_summary = city_summary.loc[city_summary["City Type"]== "Urban"]
urban_rides = urban_summary["Rides per City"]
urban_drivers = urban_summary["Drivers per City"]
urban_avg = urban_summary["Average Fare"]
#Suburban
suburban_summary = city_summary.loc[city_summary["City Type"]== "Suburban"]
suburban_rides = suburban_summary["Rides per City"]
suburban_drivers = suburban_summary["Drivers per City"]
suburban_avg = suburban_summary["Average Fare"]
#Rural
rural_summary = city_summary.loc[city_summary["City Type"]== "Rural"]
rural_rides = rural_summary["Rides per City"]
rural_drivers = rural_summary["Drivers per City"]
rural_avg = rural_summary["Average Fare"]
```

```python
# Create bubble plot
urban_plot = plt.scatter(urban_rides, urban_avg, s=urban_drivers*10, facecolors = "coral", alpha=0.75, edgecolors="black", linewidth=2, label = "Urban")
surb_plot = plt.scatter(suburban_rides, suburban_avg, s=suburban_drivers*10, facecolors = "SkyBlue", alpha=0.75, edgecolors="black", linewidth=2, label = "Suburban")
rural_plot = plt.scatter(rural_rides, rural_avg, s=rural_drivers*10, facecolors = "gold", alpha=0.75, edgecolors="black", linewidth=2, label = "Rural")
plt.xlabel("Total Number of Rides (Per City)")
plt.legend(title="City Types", handles=[urban_plot, surb_plot, rural_plot], loc="best")

plt.ylabel("Average Fare ($)" )
plt.title("Pyber Ride Sharing Data")
plt.grid()
```
