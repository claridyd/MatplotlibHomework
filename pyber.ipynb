```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Read the city and ride data
city_df = pd.read_csv("city_data.csv")
ride_df = pd.read_csv("ride_data.csv")

# City dataframe
#city_df

# Ride dataframe
#ride_df

#combine the data into a single dataset
combo_data = pd.merge(ride_df, city_df, on="city")

#Display the combo data table for preview
combo_data.head()
```

```python
# what are the city types
city_type = combo_data["type"].unique()
print(city_type)

#Group by City
city_grp = combo_data.groupby("city")


#Total number of rides per city
rides_per_city = city_grp["ride_id"].count()
#print(rides_per_city)

#Total number of drivers per city
drivers_city = city_grp["driver_count"].max()
#print(drivers_city)

#Average Fare per city
avg_fare = city_grp["fare"].mean()
#print(avg_fare)
#######################################################
# Create dataframe to hold data
#zipped_df = list(zip(rides_per_city, drivers_city, avg_fare))
#test1 = pd.DataFrame(zipped_df, columns = ["Rides per City", "Drivers per City", "Average Fare"])
#print(test1)

```
